```python
import os
import pandas as pd
```

```python
# assign directory
output_dir = os.getcwd() + '\\MOPAC_input_MNDO_sol'
xyz_dir = os.getcwd() + '\\xyz'
csv_file = os.getcwd()  # Update this to your CSV file path

# Create the new directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
print (xyz_dir)
```

```python
filtered_data_updated = pd.read_csv('filtered_data_updated_final.csv')
```

```python
# Load EPS values from CSV
eps_data = pd.read_csv('unique_solvent_names.csv')  # Assumes CSV has 'solvent_name' and 'EPS' columns
eps_dict = dict(zip(eps_data['solvent_name'], eps_data['EPS']))  # Convert to dictionary

# Iterate through solute-solvent combinations
for index, row in filtered_data_updated.iterrows():
    solute = row['solute_inchikey']
    solvent = row['solvent_name']

    # Replace spaces with underscores in the solvent name
    solvent = solvent.replace(" ", "_")

    # Construct filenames
    xyz_filename = f"{solute}.xyz"
    mop_filename = f"MNDO_{solute}_{solvent}.mop"

    full_xyz_filename = os.path.join(xyz_dir, xyz_filename)
    full_mop_filename = os.path.join(output_dir, mop_filename)

    # Ensure the solvent has an EPS value
    if solvent not in eps_dict:
        print(f"Skipping {solute} in {solvent} - EPS value not found!")
        continue

    eps_value = eps_dict[solvent]  # Get the corresponding EPS value

    # Open files and process
    if os.path.exists(full_xyz_filename):
        with open(full_xyz_filename, 'r') as xyz_file, open(full_mop_filename, 'w') as mop_file:
            # Write MOP header with correct EPS value
            mop_file.write(f'AUX LARGE OPT FORCE THERMO MNDO T=128H RECALC=5 GNORM=0.01 LET SCFCRT=0.0000001 RSOLV=1.3 EPS={eps_value}\n')
            mop_file.write(f"{solute}\n\n")  # Write solute name

            # Copy atomic coordinates from XYZ file (skip first 2 lines)
            lines = xyz_file.readlines()[2:]
            for line in lines:
                mop_file.write(line)

        print(f"Created: {full_mop_filename}")
    else:
        print(f"Missing XYZ file: {full_xyz_filename}, skipping...")
```
